# Dealing With Missing Values
 - First we check how many missing values are present

```python
kashti.isnull().sum()   # Here the missing values are in deck,age,embarked and embark_town column
```

- Second we can deal with missing values 
- we can replace missing values with Mean and Mode. Or we can remove the rows and columns 
- Preferly we can replace with mean

# Removing NAN Values

```python
# here we remove the missing values of Deck columns only
cleaned_Data=kashti.dropna(subset=['deck'],axis=0,inplace=False)
```

```python
cleaned_Data.head()
```

```python
# again checking of missing values of Deck column
cleaned_Data.isnull().sum()
```

```python
# Here we remove the remaing all missing Values 
cleaned_Data=cleaned_Data.dropna()
cleaned_Data.head()
```

- Removing values at once 

```python
# Here we make a copy of the original data set and remove the missing values at once 

ks=sns.load_dataset('Titanic')
ks.dropna().head()
```

```python
# Missing Values are removed 
cleaned_Data.isnull().sum()
```

# 2nd way of Replacing NAN values

```python
# Loading data set again for missing values 
rep=sns.load_dataset('titanic')
rep.head()
```

```python
rep.isnull().sum()
```

```python
# Here we can replace the missing values of age column with its mean values
# so first we calculate mean 

rep_mean=rep['age'].mean()
rep_mean
```

```python
# Missing Values of age have been updated with mean of age 
rep['age']=rep['age'].replace(np.nan,rep_mean)
rep.head()
```

```python
# Here the age missing values are handled
rep.isnull().sum()
```

# Here 2nd way is to deal with Catagorical Columns

```python
# checked the mode value of deck column 
rep['deck'].value_counts()
```

```python
# NAN values are filled with the mode Value
rep['deck'].fillna(rep['deck'].value_counts().index[0],inplace=True)
```

```python
rep.head()
```

```python
# NAN values of Deck column are handled
rep.isna().sum()
```

```python
# To check the mode values of embarked column 
rep['embarked'].value_counts()
```

```python
# Filled with the mode Value of embarked column 
rep['embarked'].fillna(rep['embarked'].value_counts().index[0],inplace=True)
```

```python
rep.isnull().sum()
```

```python
# To check the mode  values of embarked_town column
rep['embark_town'].value_counts()
```

```python
# Replaced with mode value 
rep['embark_town'].fillna(rep['embark_town'].value_counts().index[0],inplace=True)
```

```python
rep.isnull().sum()
```

# Data Normalization
  - Normalization is done in following methods: 
    - Features scaling method
    - Min-Max method
    - Z-score method
    - Log Transformation Method

# Features scaling Method

**Formula= column/column.max()**

```python
rep.head()
```

```python
# Here we normalize age and fare column
rep[['age','fare']].head()
```

```python
rep['age']=rep['age']/rep['age'].max()
rep['fare']=rep['fare']/rep['fare'].max()
```

```python
# Here it is in normalized form
rep[['age','fare']].head()
```

# Min-Max method
- Formula= column - column.min() / column.max() - column.min()

```python
ks=sns.load_dataset('titanic')
```

```python
# Here we again normalize age and fare column\
rep['age']-rep['age'].min()/rep['age'].max()-rep['age'].min()
rep['fare']-rep['fare'].min()/rep['fare'].max()-rep['fare'].min()
```

```python
# Yes it is normalized 
rep[['age','fare']].head()
```

# Z-score Method
  - Formula=column-column.mean()/column.std() 

```python
z=sns.load_dataset('titanic')
z.head()
```

```python
z['age']=(z['age']-z['age'].mean()) / z['age'].std()
z['fare']=(z['fare']-z['fare'].mean()) / z['fare'].std()
```

```python
z[['age','fare']].head()
```

# Log Transformation
   - Formula=np.log(data_set[column_Name])

```python
l=sns.load_dataset('titanic')
```

```python
l.head()
```

```python
# Here we again normalize the age and fare column 
l['age']=np.log(l['age'])
l['fare']=np.log(l['fare'])
```

```python
# Here it is normalized 
l[['age','fare']].head()
```
